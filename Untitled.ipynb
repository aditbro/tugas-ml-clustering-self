```python
from data_loader import load_file
from k_means import Kmeans
from dbscan import DBSCAN
from agglomerative import Agglomerative
from random import shuffle
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN as dbs
from sklearn.cluster import AgglomerativeClustering
from statistics import mode
from utils import(
    euclidean_distance,
    build_class_mapping,
    calculate_accuracy,
    build_confusion_matrix,
    show_confusion_matrix,
    build_conf_mat_from_bare
)
```

```python
raw_data = load_file('iris.data')
classes = set([x[-1] for x in raw_data])
class_dict = {}
test_data = {}
train_data = []
for kelas in classes:
    class_dict[kelas] = list(filter(lambda x: x[-1] == kelas, raw_data))
    class_dict[kelas] = list(map(lambda x: x[:-1], class_dict[kelas]))
    shuffle(class_dict[kelas])

    test_data[kelas] = class_dict[kelas][:10]
    train_data += class_dict[kelas][10:]
```

```python
kmean = Kmeans(len(classes))
kmean.fit(train_data)
mapping = build_class_mapping(class_dict, kmean)

test_result = {}
expected_result = {}
for kelas in classes:
    test_result[kelas] = [kmean.predict(x) for x in test_data[kelas]]
    expected_result[kelas] = [mapping[kelas] for x in range(len(test_data[kelas]))]

accuracy = sum([calculate_accuracy(test_result[x], expected_result[x]) for x in classes])/len(classes)
print('accuracy {}%'.format(accuracy))
print(mapping)
print("y: expected, x: predicted")
show_confusion_matrix(data=build_confusion_matrix(test_result, expected_result))
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(train_data)

mapping = build_class_mapping(class_dict, kmean)

test_result = {}
expected_result = {}
for kelas in classes:
    test_result[kelas] = [kmean.predict(x) for x in test_data[kelas]]
    expected_result[kelas] = [mapping[kelas] for x in range(len(test_data[kelas]))]

accuracy = sum([calculate_accuracy(test_result[x], expected_result[x]) for x in classes])/len(classes)
print('accuracy {}%'.format(accuracy))
print(mapping)
print("y: expected, x: predicted")
show_confusion_matrix(data=build_confusion_matrix(test_result, expected_result))
```

```python
db_scan = DBSCAN(minpts=3, epsilon=0.45)
db_scan.fit(train_data)
mapping = build_class_mapping(class_dict, db_scan)

test_result = {}
expected_result = {}
max_class_num = 0
for kelas in classes:
    test_result[kelas] = [db_scan.predict(x) for x in test_data[kelas]]
    max_class_num = max(max_class_num, max(test_result[kelas]))
    expected_result[kelas] = [mapping[kelas] for x in range(len(test_data[kelas]))]

accuracy = sum([calculate_accuracy(test_result[x], expected_result[x]) for x in classes])/len(classes)
print('accuracy {}%'.format(accuracy))
print(mapping)
print("y: expected, x: predicted")
show_confusion_matrix(data=build_confusion_matrix(test_result, expected_result, class_num=max_class_num+1))
```

```python
curr_test_data = []
expected_result = []
for i in range(len(classes)):
    kelas = list(classes)[i]
    curr_test_data += class_dict[kelas]

clustering = dbs(eps=0.4, min_samples=3).fit(curr_test_data)
clustering.labels_

created_classes = set(clustering.labels_)
mapping = {}
expected_result = []
for i in range(len(classes)):
    kelas = list(classes)[i]
    start_idx = (i)*50
    end_idx = start_idx + 50
    
    class_mode = mode(clustering.labels_[start_idx:end_idx])
    mapping[kelas] = class_mode
    expected_result += [class_mode for x in range(50)]

accuracy = calculate_accuracy(clustering.labels_, expected_result)
print('accuracy {}%'.format(accuracy))
print(mapping)
print("y: expected, x: predicted")
test_result = list(clustering.labels_)
show_confusion_matrix(data=build_conf_mat_from_bare(test_result, expected_result, created_classes))
```
